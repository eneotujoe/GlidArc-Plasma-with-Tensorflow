```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('Glidarc_Plasma_Reactor.csv')
print(df)
print("The shape of df is:", df.shape)
```

```python
time0 = df.Time.values
time = time0[:, np.newaxis]
print(time0)
print("Time array shape is:", time0.shape)
```

```python
length = df.iloc[:, 1:5].values
print(length)
print("Length array shape is:", length.shape)
```

```python
# Calculating the velocity of the gliding arc.
vel = np.divide(length, time)
print("The velocities of the gliding arc at flowrate of 40L/min, 60L/min, 800L/min, 100L/min are: \n", vel)
print("Velocity array shape is:", vel.shape)
```

```python
# Converting the numpy array to pandas dataframe and labeling the columns.
df1 = pd.DataFrame(vel, columns = ['vel_40','vel_60','vel_80','vel_100'])
df1
```

```python
# splitting a numpy array.
z1, z2, z3, z4 = np.split(df1.values, 4, axis = 1)

# unpacking Dataframe
z11, z22, z33, z44 = df1.values.T.tolist()
print(z1)
```

```python
# Concatenating the dataframes.
df2 = pd.concat([df, df1], axis = 1)
df2
```

```python
# Summary of the statistics of the data.

summary = df2.describe()
summary
```

```python
# Plotting the velocities against time.
plt.figure()
plt.scatter(df2.Time, df2.vel_40 , label = '40')
plt.scatter(df2.Time, df2.vel_60, label = '60')
plt.scatter(df2.Time, df2.vel_80 , label = '80')
plt.scatter(df2.Time, df2.vel_100, label = '100')
plt.title('Velocity against time')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.ylim(-5, 30)
plt.xlim(-0.002,0.025)
plt.show()
```
